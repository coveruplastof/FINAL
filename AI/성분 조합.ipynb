```python
# 비타민C

def check_vitamin_c_presence(product_name, dataframe, vitamin_c_ingredients):
    # 주어진 제품에 비타민C 관련 성분이 포함되어 있는지 확인합니다.
    product = dataframe[dataframe['name'] == product_name]
    if not product.empty:
        ingredients_list = eval(product['elements'].values[0])
        # Check for both full and partial matches
        return any(ingredient for ingredient in ingredients_list for vitamin_c in vitamin_c_ingredients if vitamin_c in ingredient)
    return False  # If the product is not found, return False

def recommend_products_excluding_bad_combinations(dataframe, good_combination, bad_combination):
    # 나쁜 조합의 성분을 포함하지 않고 좋은 조합의 성분을 하나라도 포함하는 제품을 추천합니다.
    recommended_products = []
    for index, row in dataframe.iterrows():
        ingredients_list = eval(row['elements'])
        # The product must not contain any bad combination ingredient
        contains_bad = any(any(bad in ingredient for ingredient in ingredients_list) for bad in bad_combination)
        # The product must contain at least one good combination ingredient
        contains_good = any(any(good in ingredient for ingredient in ingredients_list) for good in good_combination)
        if contains_good and not contains_bad:
            recommended_products.append(row['name'])
    return recommended_products

def categorize_and_recommend_products(dataframe, recommended_product_names):
    categorized_recommendations = {'에센스': [], '스킨': [], '크림': [], '로션': []}

    # Iterate over the recommended product names and categorize them based on the product type
    for product_name in recommended_product_names:
        product_row = dataframe[dataframe['name'] == product_name]
        if not product_row.empty:
            if product_row['에센스'].values[0] == 1:
                categorized_recommendations['에센스'].append(product_name)
            if product_row['스킨'].values[0] == 1:
                categorized_recommendations['스킨'].append(product_name)
            if product_row['크림'].values[0] == 1:
                categorized_recommendations['크림'].append(product_name)
            if product_row['로션'].values[0] == 1:
                categorized_recommendations['로션'].append(product_name)
                
    return categorized_recommendations

# 성분 리스트
vitamin_c_ingredients = ["비타민C", "아스코빅애씨드", "에칠아스코빌에텔", "아스코빅애씨드폴리펩타이드", "아스코빌글루코사이드"]
good_combination_ingredients = ["펩타이드", "매트릭실"]
bad_combination_ingredients = ["콜라겐", "알파하이드록시산", "살리실릭애씨드", "실리실산"]

# 사용 예시:
# 특정 제품이 비타민 C 관련 성분을 포함하고 있는지 확인
product_name_to_check = "V A 미백광채앰플"  # 실제 확인하고 싶은 제품명으로 교체해 사용하세요.
vitamin_c_present = check_vitamin_c_presence(product_name_to_check, skincare_data, vitamin_c_ingredients)

# 좋은 조합과 나쁜 조합을 바탕으로 제품 추천
recommended_product_list = recommend_products_excluding_bad_combinations(
    skincare_data, 
    good_combination_ingredients, 
    bad_combination_ingredients
)

categorized_product_recommendations = categorize_and_recommend_products(
    skincare_data, 
    recommended_product_list
)

print(vitamin_c_present)  # 해당 제품에 비타민 C 성분이 포함되어 있는지 여부 (True/False)
print(categorized_product_recommendations)  # 추천 제품 목록
```
